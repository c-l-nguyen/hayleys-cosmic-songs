```python
#Dependencies
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Pulling the csv file
bllb = "billboard_2000_2017_spotify_clean.csv"
bllb_df = pd.read_csv(bllb)
bllb_df.head(2)
```

```python
#Creating DataFrame to work and cleaning columns
bllb_df.drop(['date', 'simple_title', 'artist', 'last_pos', 'change', 'spotify_link',
               'spotify_id', 'video_link', 'genre', 'analysis_url', 'liveness', 'speechiness', 
               'acousticness', 'instrumentalness', 'time_signature', 'key', 'duration_ms', 'loudness', 
               'valence', 'mode', 'lyrics'], axis=1, inplace=True)
#Renaming columns for presentation
bllb_df.rename(columns= {"year":"Year", "title":"Track_Title", "main_artist": "Main_Artist",
                          "peak_pos": "Peak_Position", "weeks": "Weeks", "rank":"Rank", 
                          "broad_genre": "Broad_Genre", "energy":"Energy_Factor", "tempo":"Tempo", 
                          "danceability":"Danceability"}, inplace=True)
#Changing the format of the two columns needed for the analysis to numbers
bllb_df["Tempo"] = pd.to_numeric(bllb_df["Tempo"], errors='coerce')
bllb_df["Danceability"] = pd.to_numeric(bllb_df["Danceability"], errors='coerce')
bllb_df.dtypes
```

```python
#Filtering by year to group by genres on each year
b2000 = bllb_df.loc[bllb_df["Year"]==2000,:]

#Creating individual DataFrames for each year. Most likely could be accomplished by using a for loop or a more efficient way...
top_2000 = b2000.groupby(["Broad_Genre", "Year"]).agg({"Track_Title":"count", 
                                                  "Energy_Factor":"max",
                                                  "Tempo":"mean",
                                                  "Danceability":"mean"})   
```

```python
#Colapsing the information back to one single dataframe based on the measure being analyzed, working now with danceability
dance_df = pd.DataFrame({
    "Genre":["Country", "EDM", "Pop", "R&B", "Rap", "Rock", "Unknown"],
    "2000_hits": (top_2000["Danceability"].to_list()),
    "2001_hits": (top_2001["Danceability"].to_list()),
    "2002_hits": (top_2002["Danceability"].to_list()),
    "2003_hits": (top_2003["Danceability"].to_list()),
    "2004_hits": (top_2004["Danceability"].to_list()),
    "2005_hits": (top_2005["Danceability"].to_list()),
    "2006_hits": (top_2006["Danceability"].to_list()),
    "2007_hits": (top_2007["Danceability"].to_list()),
    "2008_hits": (top_2008["Danceability"].to_list()),
    "2009_hits": (top_2009["Danceability"].to_list()),
    "2010_hits": (top_2010["Danceability"].to_list()),
    "2011_hits": (top_2011["Danceability"].to_list()),
    "2012_hits": (top_2012["Danceability"].to_list()),
    "2013_hits": (top_2013["Danceability"].to_list()),
    "2014_hits": (top_2014["Danceability"].to_list()),
    "2015_hits": (top_2015["Danceability"].to_list()),
    "2016_hits": (top_2016["Danceability"].to_list()),
    "2017_hits": (top_2017["Danceability"].to_list()),
})
dance_df.head(2)
```

```python
#Setting the Genre column as an index to use on the plots
dance_df.set_index("Genre", inplace=True)
dance_df
```

```python
#Setting the axis for the plot lines, to look at genre trends over the years
years = np.arange(2000, 2018,1)
Country, = plt.plot(years, dance_df.loc["Country", dance_df.columns], color="green")
EDM, = plt.plot(years, dance_df.loc["EDM", dance_df.columns], color="blue")
Pop, = plt.plot(years, dance_df.loc["Pop", dance_df.columns], color="red")
RandB, = plt.plot(years, dance_df.loc["R&B", dance_df.columns], color="yellow")
Rap, = plt.plot(years, dance_df.loc["Rap", dance_df.columns], color="orange")
Rock, = plt.plot(years, dance_df.loc["Rock", dance_df.columns], color="purple")
Unknown, = plt.plot(years, dance_df.loc["Unknown", dance_df.columns], color="gray")

#Formatting the tickers in plot to integers
import matplotlib.ticker
locator = matplotlib.ticker.MultipleLocator(2)
plt.gca().xaxis.set_major_locator(locator)
formatter = matplotlib.ticker.StrMethodFormatter("{x:.0f}")
plt.gca().xaxis.set_major_formatter(formatter)

plt.legend(handles=[Country, EDM, Pop, RandB, Rap, Rock, Unknown], loc="best")
#Setting the axis and plot titles
plt.title("Danceability by genre over the years")
plt.xlabel("Years reviewed")
plt.ylabel("Danceability factor")
#Printing and saving plot
plt.show()
plt.savefig("Danceability factor by genre over the years")
```
