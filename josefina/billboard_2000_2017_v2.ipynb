```python
import pandas as pd
bllb = "billboard_2000_2017_spotify_clean.csv"
bllb_df = pd.read_csv(bllb)
bllb_df.head()
```

```python
#bllb_df.dtypes
bllb_df.columns
```

```python
sorted_rank = bllb_df.sort_values(by="year", ascending=True)
sorted_bllb_df = sorted_rank.sort_values(by=["year", "weeks"], ascending=False)
sorted_bllb_df.head()
```

```python
def top_10(values):
    return list(values[0:10])
```

```python
df = sorted_bllb_df.groupby("year").agg({
    "main_artist": top_10,
    "weeks": top_10,
    "rank": top_10,
    "broad_genre": top_10,
    "danceability": top_10})
df.head()
```

```python
df["combined"] = df.apply(lambda row: list(zip(row["artist"], row["rank"], )), axis=1)
df.head()
```

```python
df.explode("combined")
```
