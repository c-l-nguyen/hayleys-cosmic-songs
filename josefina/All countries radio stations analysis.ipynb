```python
import pandas as pd
all_stations = "All_countries_radio_stations_data.csv"
all_stations_df = pd.read_csv(all_stations)
all_stations_df.head()
```

```python
all_stations_count = all_stations_df["Country"].value_counts()
all_stations_count
```

```python
france = all_stations_df.loc[all_stations_df["Country"]=="FR",:]

france_counts= france.groupby("Genre", as_index=False).agg({"Radio ID":"count"})
france_counts= france_counts.sort_values("Radio ID", ascending=False)
france_counts
top_fr = france_counts.head(5)
print(top_fr)
top_fr.to_csv("French_radio_stations_genres", header=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
labels = top_fr["Genre"]
colors = ["gold", "lightcoral", "lightskyblue", "green", "orange"]
explode = (0,0,0,0,0.15)
plot = top_fr.plot.pie(y="Radio ID", explode=explode, labels=labels, autopct="%1.1f%%", figsize=(5,5), shadow=True, 
                   colors=colors, startangle=90)
plt.title("Top 5 genres in French radio stations")
plt.ylabel("")
plt.savefig("Top_5_France_chart")
plt.show()
```
