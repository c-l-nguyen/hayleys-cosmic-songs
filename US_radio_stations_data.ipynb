```python
import requests
import json

api_key = #"enter key here"
url = "https://30-000-radio-stations-and-music-charts.p.rapidapi.com/rapidapi"
radio = url + "/" + api_key + "?countries="
```

```python
keyword ="ALL"
country = "US"
querystring = {"country":country,"keyword":keyword,"genre":"ALL"}

headers = {
    'x-rapidapi-host': "30-000-radio-stations-and-music-charts.p.rapidapi.com",
    'x-rapidapi-key': api_key
    }
response = requests.get(url, headers=headers, params=querystring).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
import pandas as pd
#from pandas.io.json import json_normalize
results = response["results"]
#print(results)
# response_df = pd.DataFrame.from_dict(json_normalize(response), orient='columns')
response_df = pd.DataFrame.from_dict(results, orient='columns')
response_df.head(5)
```

```python
response_df.columns
```

```python
response_df.rename(columns = {"c":"Country", "g":"Genre", "i":"Radio ID", "l":"Radio logo", "n":"Radio Name", 
                              "u":"Stream URL"}, inplace=True)
response_df.drop(["d"], axis=1, inplace=True)
response_df.columns
```

```python
response_df.to_csv("Radio_stations_data.csv", encoding="utf-8", index=False, header=True)
```
