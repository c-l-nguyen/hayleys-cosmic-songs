```python
import requests
import json

api_key = "6eed7eabfeadf7e13784e1afec29ed52"
url = "http://ws.audioscrobbler.com/2.0/"   #?method=chart.gettoptags"

method = "chart.gettoptracks"
params = {
    "method": method,
    "api_key": api_key,
    "format": "json"
}

tracks = requests.get(url, params=params)
tracks_json = tracks.json()
print(tracks.url)
```

```python
tracks_response = tracks_json["tracks"]["track"]
tracks_response
```

```python
import pandas as pd

tracks_response_df = pd.DataFrame(tracks_response)
tracks_response_df.head()
```
