```python
import spotipy
import time
from spotipy.oauth2 import SpotifyClientCredentials
import pprint
import pandas as pd

pp = pprint.PrettyPrinter(indent=1)

client_credentials_manager = SpotifyClientCredentials(client_id="4ec0a62d437c4ce787a498c11673e555",
                                                      client_secret="81db6bb3fcc041aa8e2b06cbc19e9a91")
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
# read in the data
top_100_df = pd.read_csv("../data/clean/billboard_top_100_by_decade.csv")
top_100_df["main_performer"] = top_100_df["performer"].str.split("Featuring").apply(lambda x: x[0])
```

```python
# because requests limit is dynamically determined, just manually edit ranges until done...
for index, row in top_100_df.iloc[400:].iterrows():
    artist = row["main_performer"]
    track = row["song"]
    query = f"artist:{artist} track:{track}"

    results = sp.search(q=query, type='track')["tracks"]["items"]

    if len(results) > 0:
        main_result = results[0]
        release_date = main_result["album"]["release_date"]
        title = main_result["name"]
        main_artist = main_result["album"]["artists"][0]["name"]
        spotify_id = main_result["id"]
        spotify_link = main_result["uri"]
        popularity = main_result["popularity"]
        explicit = main_result["explicit"]

        top_100_df.loc[index, "release_date"] = release_date
        top_100_df.loc[index, "title"] = title
        top_100_df.loc[index, "main_artist"] = main_artist
        top_100_df.loc[index, "spotify_id"] = spotify_id
        top_100_df.loc[index, "spotify_link"] = spotify_link
        top_100_df.loc[index, "popularity"] = popularity
        top_100_df.loc[index, "explicit"] = explicit
    else:
        top_100_df.loc[index, "release_date"] = ""
        top_100_df.loc[index, "title"] = ""
        top_100_df.loc[index, "main_artist"] = ""
        top_100_df.loc[index, "spotify_id"] = ""
        top_100_df.loc[index, "spotify_link"] = ""
        top_100_df.loc[index, "popularity"] = ""
        top_100_df.loc[index, "explicit"] = ""

top_100_df
```

```python
# save off the records with missing data. To be investigated later
missing_df = top_100_df.query("title == ''")
missing_df
```

```python
# get the audio features of each track with an existing spotify_id
for index, row in top_100_df.iloc[400:].iterrows():
    spotify_id = row["spotify_id"]
    
    if spotify_id:
        audio_features = sp.audio_features(spotify_id)[0]
        
        top_100_df.loc[index,"acousticness"] = audio_features["acousticness"]
        top_100_df.loc[index,"analysis_url"] = audio_features["analysis_url"]
        top_100_df.loc[index,"danceability"] = audio_features["danceability"]
        top_100_df.loc[index,"duration_ms"] = audio_features["duration_ms"]
        top_100_df.loc[index,"energy"] = audio_features["energy"]
        top_100_df.loc[index,"instrumentalness"] = audio_features["instrumentalness"]
        top_100_df.loc[index,"key"] = audio_features["key"]
        top_100_df.loc[index,"liveness"] = audio_features["liveness"]
        top_100_df.loc[index,"loudness"] = audio_features["loudness"]
        top_100_df.loc[index,"mode"] = audio_features["mode"]
        top_100_df.loc[index,"speechiness"] = audio_features["speechiness"]
        top_100_df.loc[index,"tempo"] = audio_features["tempo"]
        top_100_df.loc[index,"time_signature"] = audio_features["time_signature"]
        top_100_df.loc[index,"valence"] = audio_features["valence"]
    
top_100_df
```

```python
# get the genres of each artist
for index, row in top_100_df.iloc[400:].iterrows():
    artist = row["main_artist"]

    if artist:
        genres = sp.search(q=artist, type='artist')["artists"]["items"][0]["genres"]
        top_100_df.loc[index,"genres"] = str(genres) # convert to string since can't save list object in DataFrame

top_100_df
```

Now, you may be wondering why I split up these 3 calls into three separate loops even though I could have done one loop. It's entirely due to the dynamically determined rate limit for the Spotify API. If I had put all 3 calls into one loop, then I would have hit the server more often and would be more limited in how many rows I could get for each loop. Instead, I decided that I could just split into 3 loops, 1 for each call, and inspect partial results each time.
